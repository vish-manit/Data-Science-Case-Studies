```python
import nltk
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('omw-1.4')
import numpy as np
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem.wordnet import WordNetLemmatizer
from ast import literal_eval
```

```python
df = pd.read_csv("goibibo_com-travel_sample.csv")
df.head()
```

```python
df.drop(['additional_info', 'area', 'country', 'crawl_date', 'hotel_brand','hotel_description','room_count', 'room_facilities', 'room_type','similar_hotel', 'site_review_count','site_stay_review_rating', 'sitename', 'uniq_id'], axis = 1, inplace = True)
```

```python
df.drop(['hotel_facilities', 'image_count', 'latitude', 'longitude', 'locality', 'property_id', 'property_name', 'property_type', 'province', 'qts', 'query_time_stamp', 'review_count_by_category', 'room_area'], axis = 1, inplace=True)
df.head()
```

```python
df.drop(['pageurl', 'point_of_interest'], axis = 1, inplace = True)
df.head()
```

```python
def impute(column):
    column = column[0]
    if (type(column) != list):
        return "".join(literal_eval(column))
    else:
        return column
    
```

```python
df['city'] = df['city'].str.lower()
df['state'] = df['state'].str.lower()
df.head()
```

```python
def recommend_hotel(location, description):
    description = description.lower()
    description_tokens = word_tokenize(description)
    stop_words = stopwords.words('english')
    lemm = WordNetLemmatizer()
    filtered = {word for word in description_tokens if not word in stop_words}
    filtered_set = set()
    for fs in filtered:
        filtered_set.add(lemm.lemmatize(fs))

    city = df[df['city'] == location.lower()]
    city = city.set_index(np.arange(city.shape[0]))
    list1 = []
    list2 = []
    cos = []
    for i in range(city.shape[0]):
        temp_token = word_tokenize(city["hotel_category"][i])
        temp_set = [word for word in temp_token if not word in stop_words]
        temp2_set = set()
        for s in temp_set:
            temp2_set.add(lemm.lemmatize(s))
        vector = temp2_set.intersection(filtered_set)
        cos.append(len(vector))
    city['similarity'] = cos
    city = city[city['hotel_category'].str.contains('gostays', case=False)]  # Filter by hotel category
    city = city.sort_values(by='similarity', ascending=False)
    city.drop_duplicates(subset='address', keep='first', inplace=True)
    city.sort_values('site_review_rating', ascending=False, inplace=True)
    city.reset_index(inplace=True)
    return city[["address", "site_review_rating", "hotel_category"]].head()

```

```python
recommend_hotel('gurgaon', 'gostays')
```
